Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
#solutions:- 
import pandas as pd
def re_index(df):
    
    new_index = pd.Series(range(1,len(df)*2,2))
    
    df.index = new_index
    
    return df
```

```python
df = pd.DataFrame({'A': [1,2,3],'B':[4,5,6],'C':[7,8,9]})
```

```python
df_new = re_index(df)
```

```python
print(df_new)
```

Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The function should print the sum to the console.For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should
calculate and print the sum of the first three values, which is 60.

```python
#solution
import pandas as pd

def calculate_sum_first(df):
    #Ensure values column exist in the DataFrame
    if 'Values' not in df.columns:
        print("Error: Value column not found in the data frame")
        return 
    
    #Extract the first three values from the 'Values' column 
    first_three_values = df['Values'].head(3)
    
    
    sum_of_values = first_three_values.sum()
    
    print(f"Sum of first three values :{sum_of_values}")
    


df1 = pd.DataFrame({'Values':[10,20,30,40,50]})

calculate_sum_first(df1)
```

addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address.
The username is the part of the email address that appears before the '@' symbol. For example, if the
email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your
function should extract the username from each email address and store it in the new 'Username'
column.

```python
#solution
import pandas as pd
df = pd.DataFrame({'Email':['singhraja@gmail.com','mundakukku@gmail.com','oyepape@gmail.com']})
```

Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the values in the 'Values' column.

```python
#solution
import pandas as pd
df = pd.DataFrame({"Values":[10,20,30,40,50]})
```

```python
def fun(df):
    mean_value = df['Values'].mean()
    
    std_devaition = df['Values'].std()
    
    median = df['Values'].median()
    
    return mean_value,std_devaition,median
    
```

```python
mean_val,std_val,median_val = fun(df)
print(f'The  mean  is {mean_val}')
print(f'The  mean  is {std_val}')
print(f'The  mean  is {median_val}')
```

Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to
create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days
for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
should include the current day.

```python
#Solution
import pandas as pd
df = pd.DataFrame({'Sales':[100,200,300,50,60,350,500,100,300,30,150,160,30,50,300,310,50,60,30,510],"Date":pd.date_range(start = '10-06-24',periods = 20)})
```

```python
def calculating_moving_average(df,window_size = 7):
    
    df['MovingAverage'] = df['Sales'].rolling(window = window_size, min_periods = 1).mean()
    
    return df
```

```python
calculating_moving_average(df)
```

Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new
column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.
Monday, Tuesday) corresponding to each date in the 'Date' column.
For example, if df contains the following values:
Date
0 2023-01-01
1 2023-01-02
2 2023-01-03
3 2023-01-04
4 2023-01-05
Your function should create the following DataFrame:

Date Weekday
0 2023-01-01 Sunday
1 2023-01-02 Monday
2 2023-01-03 Tuesday
3 2023-01-04 Wednesday
4 2023-01-05 Thursday
The function should return the modified DataFrame.

```python
#solution
import pandas as pd
df = pd.DataFrame({'Date':[' 2023-01-01','2023-01-02' , '2023-01-03' , '2023-01-04','2023-01-05']})
```

```python
df.dtypes
```

```python
def new_column(df):
    
    #covert the date column to datetime if it is not already 
    df['Date'] = pd.to_datetime(df['Date'])
    
    #Adding new column weekday which contain the weekdays name 
    df['Weekday'] = df['Date'].dt.day_name()
    
    return df
```

```python
new_column(df)
```

Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
#solution
import pandas as pd
df = pd.DataFrame({'Date':['2023-01-01','2023-12-10','2023-04-15','2023-01-20','2023-01-31','2023-01-21']})
```

```python
def select_rows_by_date_range(df):
    
    #Ensure the 'Date' column is in the datetime format 
    df['Date'] = pd.to_datetime(df['Date'])
    
    start_date = '2023-01-01'
    end_date = '2023-01-31'
    
    
    filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
    
    return filtered_df
    
```

```python
select_rows_by_date_range(df)
```
